```python
import pandas as pd
import yfinance as yf

# load excel file
acoes = pd.read_excel('posicao.xlsx', sheet_name="Ações")
acoes.dropna(inplace=True)
etf = pd.read_excel('posicao.xlsx', sheet_name="ETF")
etf.dropna(inplace=True)
fundos = pd.read_excel('posicao.xlsx', sheet_name="Fundo de Investimento")
fundos.dropna(inplace=True)
tesouro = pd.read_excel('posicao.xlsx', sheet_name="Tesouro Direto")
tesouro.dropna(inplace=True)
```

```python
# ações/fundos
acoes_and_fundos = acoes.append(fundos)
acoes_and_fundos .sort_values("Código de Negociação", inplace=True)
prices = yf.download(tickers=(acoes_and_fundos ["Código de Negociação"]+'.SA').to_list(), threads=False)
prices = prices['Adj Close'].tail(2)    
prices.columns  = prices.columns.str.rstrip('.SA')
prices = prices.transpose()
prices.columns = ["LastDay_value", "Day_value"]
prices["Quantidade"] = (acoes_and_fundos ["Quantidade"]).to_list()
prices["Quantidade"] = prices["Quantidade"].astype(int)
prices["Total"] = (prices["Quantidade"] * prices["Day_value"]).round(2)
prices["Return"] = ((prices["Day_value"] - prices["LastDay_value"]) * prices["Quantidade"]).round(2)
prices["Pct"] = (((prices["Day_value"] / prices["LastDay_value"])-1) *100).round(2) 
print("\n[", 20*"*", "AÇÕES/FUNDOS IMOBILIÁRIOS", 20*"*", "]\nRETORNO DO DIA ==> ", prices["Return"].sum().round(2))
prices
```
